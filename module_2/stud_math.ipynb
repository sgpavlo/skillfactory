Самый простой способ отфильтровать выбросы — воспользоваться формулой межквартильного размаха. Выбросом считаются такие значения, которые лежат вне этого промежутка

[percentile(25) -1.5IQR : percentile(75)+1.5IQR]

где IQR = percentile(75) - percentile(25). Для этого создадим функцию def outlier_filter(col).

Функция для определения границ выбросов

```python
def outlier_filter(col):
    perc25 = stud.loc[:, col].quantile(0.25)
    perc75 = stud.loc[:, col].quantile(0.75)
    IQR = perc75 - perc25
    left = perc25 - 1.5*IQR
    right = perc75 + 1.5*IQR
    return [left, right]
```

```python
# Использовать при необходимости, если решим избавиться от всех выбросов
#outlier_int = num_value
```

Удалим только значение '-1' в колонке 'famrel' и 'Fedu' значение 40. Другие выбросы мы всегда сможем удалить при необходимости использовав функцию  для очистки от выбросов.

```python
def outlier_change_nan(list):
    for out in list:
        border = outlier_filter(out)
        stud.loc[:, out] = stud.loc[:, out].apply(
            lambda x: None if x > border[1] else x)
        stud.loc[:, out] = stud.loc[:, out].apply(
            lambda x: None if x < border[0] else x)
```

Один из этих столбцов можно просто удалить либо (для тренировки) выполним взаимозаполнения пустых значений в этих столбцах.

Создадим функции def fill_fedu(row), def fill_medu(row), которые будут заполнять пустые значение 'Fedu' и 'Medu'.

```python
#заполняет пустые значение 'Fedu'
def fill_fedu(row):
    if pd.isnull(row['Fedu']):
        if row['Medu'] == 0:
            return 0
        if row['Medu'] == 1:
            return 1
        if row['Medu'] == 2:
            return 2
        if row['Medu'] == 3:
            return 3
        return 4
    return row['Fedu']
```

```python
#заполняет пустые значение 'Medu'
def fill_medu(row):
    if pd.isnull(row['Medu']):
        if row['Fedu'] == 0:
            return 0
        if row['Fedu'] == 1:
            return 1
        if row['Fedu'] == 2:
            return 2
        if row['Fedu'] == 3:
            return 3
        return 4
    return row['Medu']
```

```python
stud['Fedu'] = stud.apply(fill_fedu, axis=1)
stud['Medu'] = stud.apply(fill_medu, axis=1)
```

Судя по графикам в некоторых параметрах тяжело определить влияние на балы по госэкзамену по матиматике(score). Графики являются лишь вспомогательным инструментом, настоящую значимость различий может помочь распознать статистика. Проверим, есть ли статистическая разница в распределении оценок по номинативным признакам, с помощью теста Стьюдента. Проверим нулевую гипотезу о том, что распределения оценок по различным параметрам неразличимы: Для этого создадим функцию stat_different, которая выведет статистически значимые различия для колонок.

```python
#Ищем статистически значимые различия для колонки с номинативными переменными
def get_stat_dif(column):
    cols = stud.loc[:, column].value_counts().index[:10]
    combinations_all = list(combinations(cols, 2))
    for comb in combinations_all:
        if ttest_ind(stud.loc[stud.loc[:, column] == comb[0], 'score'],
                     stud.loc[stud.loc[:, column] == comb[1], 'score']).pvalue \
                <= 0.05/len(combinations_all):  # Учли поправку Бонферони
            print('Найдены статистически значимые различия для колонки', column)
            break
```

```python
# Пройдемся по столбцам применив функцию stat_different.
for col in ['school', 'sex', 'address', 'famsize', 'Pstatus', 'Mjob', 'Fjob',
            'reason', 'guardian', 'schoolsup', 'famsup', 'paid', 'activities',
            'nursery', 'higher', 'internet', 'romantic']:
    get_stat_dif(col)
```
